```python
import os #Librería funciones del Sistema Operativo
import time #Librería para el tiempo

E_NINO = 2500 #Valor Entrada niño
C_NINO = 0 #Cantidad Entrada niño a comprar
AC_NINO = 0 #Acumulador Entrada niño
T_NINO = 0 #Total Entrada niño
E_ADULTO = 5000 #Valor Entrada adulto
C_ADULTO = 0 #Cantidad Entrada adulto a comprar
AC_ADULTO = 0 #Acumulador Entrada adulto
T_ADULTO = 0 #Total Entrada Adulto
E_ADULTO_M = 1000 #Valor Entrada adulto mayor
AC_ADULTO_M = 0 #Acumulador Entrada adulto mayor
C_ADULTO_M = 0 #Cantidad Entrada adulto mayor a comprar
T_ADULTO_M = 0 #Total Entrada Adulto
Flag = True #variable de control para el ciclo de repetición
Flag2 = True #variable de control segundo ciclo
SUBTOTAL = 0 #Variable sub total
TOTAL = 0 #variable acumulativa
DESCUENTO_N = 0 #Descuento niño
DESCUENTO_M = 0 #Descuento Adulto
DESCUENTO = 0 #Descuento Total
op = 0 #Primera opciones
op2 = 0 #Segunda opciones
print("___________________________________________________")
print("|  Tipo Entrada  |     Detalle       |   Precio   |")
print("___________________________________________________")
print("|  Menores       | De 5 a 12 años    | $2.500     |")
print("___________________________________________________")
print("|  Adultos       | De 13 a 64 años   | $5.000     |")
print("___________________________________________________")
print("|  Adultos       | Desde 65 y más    |            |")
print("|  Mayores       |                   | $1.000     |")
print("___________________________________________________")
while Flag:
  print("\n---------------------------------------------")
  print("\n                Menú Principal               ")
  print("\n---------------------------------------------")
  print("\n\t 1. Tipo Entrada\n\t 2. Pago\n\t 3. Mostrar compra\n\t 4. Salir")
  try:
    op = int(input("Opción:\n\t"))
    if (op > 0 and op < 5):
      if op == 1: #ENTRADAS
        while Flag2:
          print("\n Tipo de Entrada\n\t 1. Menores (5 a 12 años)\n\t 2. Adultos (13 a 64 años)\n\t 3. Adultos Mayores (Desde 65 años y más)\n\t 4. Salir Sub-Menú\n\t")
          op2 = int(input("Opción:\n\t"))
          if (op2 >= 1 and op2 <=4):
            if op2 == 1:
              print("\n Ha seleccionado la opción 1\n\tCuantas Entradas de Niño desea?: ")
              C_NINO = int(input("Cantidad:\n\t"))
              AC_NINO = AC_NINO + C_NINO
            elif op2 == 2:
              print("\n Ha seleccionado la opción 2\n\tCuantas Entradas de Adulto desea?: ")
              C_ADULTO = int(input("Cantidad:\n\t"))
              AC_ADULTO = AC_ADULTO + C_ADULTO
            elif op2 == 3:
              print("\n Ha seleccionado la opción 3\n\tCuantas Entradas de Adulto Mayor desea?: ")
              C_ADULTO_M = int(input("Cantidad:\n\t"))
              AC_ADULTO_M = AC_ADULTO_M + C_ADULTO_M
            elif op2 == 4:
              print("\n Ha seleccionado la opción 4\n\tRegresará al Menú Principal\n\t")
              Flag2 = False
          else:
            print("\n Ha ingresado una opción invalida\n\t")
      elif op == 2: #Pago
        print("\n Ingrese que día de la semana es: \n\t 1. LUNES\n\t 2. MARTES\n\t 3. MIÉRCOLES\n\t 4. JUEVES\n\t 5. VIERNES\n\t 6. SÁBADO\n\t 7. DOMINGO\n\t")
        op2 = int(input("Opción:\n\t"))
        if (op2 >= 1 and op2 <=7):
          if op2 == 1: #DIA LUNES, PUEDE AGREGAR DESCUENTOS SI LO DESEA
            T_NINO = AC_NINO * E_NINO
            T_ADULTO = AC_ADULTO * E_ADULTO
            T_ADULTO_M = AC_ADULTO_M * E_ADULTO_M
            SUBTOTAL = T_NINO + T_ADULTO + T_ADULTO_M
            DESCUENTO_N = 0
            DESCUENTO_A = 0
            DESCUENTO = DESCUENTO_N + DESCUENTO_A
            TOTAL = SUBTOTAL - DESCUENTO
            print("\n ES LUNES: ")
            print("\n No hay descuento vigente \n\n")
            print("Entradas")
            print("--------------------------------------------")
            print(f"● {AC_NINO} Entradas Menores          ${T_NINO}")
            print(f"● {AC_ADULTO} Entradas Adultos          ${T_ADULTO}")
            print(f"● {AC_ADULTO_M} Entradas Adultos Mayores  ${T_ADULTO_M}")
            print("--------------------------------------------")
            print(f"Subtotal                        ${SUBTOTAL}")
            print(f"Descuentos                      ${DESCUENTO}")
            print("--------------------------------------------")
            print(f"Total a pagar                   ${TOTAL}")
            print("\n\n            'Gracias por su Compra'")

          elif op2 == 2: #DIA MARTES, PUEDE AGREGAR DESCUENTOS SI LO DESEA
            T_NINO = AC_NINO * E_NINO
            T_ADULTO = AC_ADULTO * E_ADULTO
            T_ADULTO_M = AC_ADULTO_M * E_ADULTO_M
            SUBTOTAL = T_NINO + T_ADULTO + T_ADULTO_M
            DESCUENTO_N = 0
            DESCUENTO_A = 0
            DESCUENTO = DESCUENTO_N + DESCUENTO_A
            TOTAL = SUBTOTAL - DESCUENTO
            print("\n ES MARTES: ")
            print("\n No hay descuento vigente \n\n")
            print("Entradas")
            print("--------------------------------------------")
            print(f"● {AC_NINO} Entradas Menores          ${T_NINO}")
            print(f"● {AC_ADULTO} Entradas Adultos          ${T_ADULTO}")
            print(f"● {AC_ADULTO_M} Entradas Adultos Mayores  ${T_ADULTO_M}")
            print("--------------------------------------------")
            print(f"Subtotal                        ${SUBTOTAL}")
            print(f"Descuentos                      ${DESCUENTO}")
            print("--------------------------------------------")
            print(f"Total a pagar                   ${TOTAL}")
            print("\n\n            'Gracias por su Compra'")

          elif op2 == 3: #DIA MIÉRCOLES, PUEDE AGREGAR DESCUENTOS SI LO DESEA
            T_NINO = AC_NINO * E_NINO
            T_ADULTO = AC_ADULTO * E_ADULTO
            T_ADULTO_M = AC_ADULTO_M * E_ADULTO_M
            SUBTOTAL = T_NINO + T_ADULTO + T_ADULTO_M
            DESCUENTO_N = 0
            DESCUENTO_A = 0
            DESCUENTO = DESCUENTO_N + DESCUENTO_A
            TOTAL = SUBTOTAL - DESCUENTO
            print("\n ES MIERCOLES: ")
            print("\n No hay descuento vigente \n\n")
            print("Entradas")
            print("--------------------------------------------")
            print(f"● {AC_NINO} Entradas Menores          ${T_NINO}")
            print(f"● {AC_ADULTO} Entradas Adultos          ${T_ADULTO}")
            print(f"● {AC_ADULTO_M} Entradas Adultos Mayores  ${T_ADULTO_M}")
            print("--------------------------------------------")
            print(f"Subtotal                        ${SUBTOTAL}")
            print(f"Descuentos                      ${DESCUENTO}")
            print("--------------------------------------------")
            print(f"Total a pagar                   ${TOTAL}")
            print("\n\n            'Gracias por su Compra'")

          elif op2 == 4: #DIA JUEVES, PUEDE AGREGAR DESCUENTOS SI LO DESEA
            T_NINO = AC_NINO * E_NINO
            T_ADULTO = AC_ADULTO * E_ADULTO
            T_ADULTO_M = AC_ADULTO_M * E_ADULTO_M
            SUBTOTAL = T_NINO + T_ADULTO + T_ADULTO_M
            DESCUENTO_N = 0
            DESCUENTO_A = 0
            DESCUENTO = DESCUENTO_N + DESCUENTO_A
            TOTAL = SUBTOTAL - DESCUENTO
            print("\n ES JUEVES: ")
            print("\n No hay descuento vigente \n\n")
            print("Entradas")
            print("--------------------------------------------")
            print(f"● {AC_NINO} Entradas Menores          ${T_NINO}")
            print(f"● {AC_ADULTO} Entradas Adultos          ${T_ADULTO}")
            print(f"● {AC_ADULTO_M} Entradas Adultos Mayores  ${T_ADULTO_M}")
            print("--------------------------------------------")
            print(f"Subtotal                        ${SUBTOTAL}")
            print(f"Descuentos                      ${DESCUENTO}")
            print("--------------------------------------------")
            print(f"Total a pagar                   ${TOTAL}")
            print("\n\n            'Gracias por su Compra'")

          elif op2 == 5: #DIA VIERNES, PUEDE AGREGAR MAS DESCUENTOS SI LO DESEA
            T_NINO = AC_NINO * E_NINO
            T_ADULTO = AC_ADULTO * E_ADULTO
            T_ADULTO_M = AC_ADULTO_M * E_ADULTO_M
            SUBTOTAL = T_NINO + T_ADULTO + T_ADULTO_M
            DESCUENTO_N = T_NINO * 0.1
            DESCUENTO_A = T_ADULTO * 0.05
            DESCUENTO = DESCUENTO_N + DESCUENTO_A
            TOTAL = SUBTOTAL - DESCUENTO
            print("\n ES VIERNES: ")
            print("\n Menores y Adultos Poseen Descuento \n\n")
            print("Entradas")
            print("--------------------------------------------")
            print(f"● {AC_NINO} Entradas Menores          ${T_NINO}")
            print(f"● {AC_ADULTO} Entradas Adultos          ${T_ADULTO}")
            print(f"● {AC_ADULTO_M} Entradas Adultos Mayores  ${T_ADULTO_M}")
            print("--------------------------------------------")
            print(f"Subtotal                        ${SUBTOTAL}")
            print(f"Descuentos                      ${round(DESCUENTO)}")
            print("--------------------------------------------")
            print(f"Total a pagar                   ${round(TOTAL)}")
            print("\n\n            'Gracias por su Compra'")

          elif op2 == 6: #DIA SABADO, PUEDE AGREGAR DESCUENTOS SI LO DESEA
            T_NINO = AC_NINO * E_NINO
            T_ADULTO = AC_ADULTO * E_ADULTO
            T_ADULTO_M = AC_ADULTO_M * E_ADULTO_M
            SUBTOTAL = T_NINO + T_ADULTO + T_ADULTO_M
            DESCUENTO_N = 0
            DESCUENTO_A = 0
            DESCUENTO = DESCUENTO_N + DESCUENTO_A
            TOTAL = SUBTOTAL - DESCUENTO
            print("\n ES SÁBADO: ")
            print("\n No hay descuento vigente \n\n")
            print("Entradas")
            print("--------------------------------------------")
            print(f"● {AC_NINO} Entradas Menores          ${T_NINO}")
            print(f"● {AC_ADULTO} Entradas Adultos          ${T_ADULTO}")
            print(f"● {AC_ADULTO_M} Entradas Adultos Mayores  ${T_ADULTO_M}")
            print("--------------------------------------------")
            print(f"Subtotal                        ${SUBTOTAL}")
            print(f"Descuentos                      ${DESCUENTO}")
            print("--------------------------------------------")
            print(f"Total a pagar                   ${TOTAL}")
            print("\n\n            'Gracias por su Compra'")
          
          elif op2 == 7: #DIA DOMINGO, PUEDE AGREGAR DESCUENTOS SI LO DESEA
            T_NINO = AC_NINO * E_NINO
            T_ADULTO = AC_ADULTO * E_ADULTO
            T_ADULTO_M = AC_ADULTO_M * E_ADULTO_M
            SUBTOTAL = T_NINO + T_ADULTO + T_ADULTO_M
            DESCUENTO_N = 0
            DESCUENTO_A = 0
            DESCUENTO = DESCUENTO_N + DESCUENTO_A
            TOTAL = SUBTOTAL - DESCUENTO
            print("\n ES DOMINGO: ")
            print("\n No hay descuento vigente \n\n")
            print("Entradas")
            print("--------------------------------------------")
            print(f"● {AC_NINO} Entradas Menores          ${T_NINO}")
            print(f"● {AC_ADULTO} Entradas Adultos          ${T_ADULTO}")
            print(f"● {AC_ADULTO_M} Entradas Adultos Mayores  ${T_ADULTO_M}")
            print("--------------------------------------------")
            print(f"Subtotal                        ${SUBTOTAL}")
            print(f"Descuentos                      ${DESCUENTO}")
            print("--------------------------------------------")
            print(f"Total a pagar                   ${TOTAL}")
            print("\n\n            'Gracias por su Compra'")
        else:
          print("\n Ha ingresado una opcion invalida\n\t")
      elif op == 3: #MOSTRAR COMPRA
        print(f"\n Mostrar Compra\n\t ● {AC_NINO} Entradas Menores\n\t ● {AC_ADULTO} Entradas Adultos\n\t ● {AC_ADULTO_M} Entradas Adultos Mayores")
      elif op == 4: #SALIR
        print(" Ha salido del Menú")
        Flag = False
    else:
      print("La opción ingresada no es correcta, intentelo nuevamente")
  except ValueError:
    print("Ingreso erroneo")
    continue
  except TypeError:
    print("Resultado erroneo")
    continue
  finally:
    if os.name == "ce" or os.name == "nt" or os.name == "dos": #Identificar el SO == Windows
      time.sleep(5)
      os.system("cls") #Limpiar la consola
      continue
    else:
      time.sleep(5)
      os.system("clear") #Limpiar la consola en los SO Linux o Ios
      continue
```
